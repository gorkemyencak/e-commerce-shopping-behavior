```python
import pandas as pd
import sys
from pathlib import Path

projectRoot = Path().resolve().parent
sys.path.append(str(projectRoot))
from src.data_loader import KaggleCSVLoader
```

```python
# initializing the loader class
datasetName = 'dhrubangtalukdar/e-commerce-shopper-behavior-amazonshopify-based'
csvFilename = 'e_commerce_shopper_behaviour_and_lifestyle.csv'
rawDataDir = projectRoot / 'data' / 'raw'

loader = KaggleCSVLoader(
    dataset_name=datasetName,
    csv_filename=csvFilename,
    raw_data_dir=rawDataDir
)

rawdata_df = loader.run()
print(rawdata_df.shape)

raw_path = projectRoot / 'data' / 'raw' / csvFilename
rawdata_df.to_csv(raw_path, index = False)
```
